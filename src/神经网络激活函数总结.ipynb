```python
import random
import math
import operator as op
from functools import reduce
from random import randint
import numpy as np
from matplotlib import pyplot as plt

# 原创
```

```python
# 逻辑函数Sigmoid
def sigmoid(x):
    return 1. / (1. + np.exp(-x))


x = np.arange(-10, 10, 0.2)
y = sigmoid(x)
plt.plot(x, y, label=r'$\Phi(x)=Sigmoid(x)=\frac{1}{1+e^{-x}}$')

plt.legend()
plt.savefig('../images/激活函数/sigmoid.png')
plt.show()
```

```python
# 正切函数Tanh
def tanh(x):
    return (1. - np.exp(-2 * x)) / (1. + np.exp(-2 * x))


x = np.arange(-10, 10, 0.2)
y = tanh(x)
plt.plot(x, y, label=r'$\Phi(x)=tanh(x)=\frac{1-e^{-2x}}{1+e^{-2x}}$')

plt.legend()
plt.savefig('../images/激活函数/tanh.png')
plt.show()
```

```python
# 线性整流函数ReLU
def relu(x):
    return np.where(x<0,0,x)


x = np.arange(-10, 10, 0.2)
y = relu(x)
plt.plot(x, y, label=r'$\Phi(x)=ReLU(x)=max(0,x)$')

plt.legend()
plt.savefig('../images/激活函数/relu.png')
plt.show()
```

```python
# ELU函数
def elu(x):
    return np.where(x>=0,x, np.exp(x)-1)


x = np.arange(-10, 10, 0.2)
y = elu(x)
#  \Phi(\alpha,x)=ELU(\alpha,x)=\begin{cases}{\alpha(e^{x}-1),x<0}\\{x,x\geq{0}}\end{cases}
plt.plot(x, y, label=r'$\Phi(\alpha,x)=ELU(\alpha,x)$')

plt.legend()
plt.savefig('../images/激活函数/elu.png')
plt.show()
```

```python
# Softplus函数
def softplus(x):
    return np.log(1+np.exp(x))


x = np.arange(-10, 10, 0.2)
y = softplus(x)
plt.plot(x, y, label=r'$\Phi(x)=ln(1+e^x)$')

plt.legend()
plt.savefig('../images/激活函数/softplus.png')
plt.show()
```

```python
# 恒同映射Identity
def identity(x):
    return x


x = np.arange(-10, 10, 0.2)
y = identity(x)
plt.plot(x, y, label=r'$\Phi(x)=x$')

plt.legend()
plt.savefig('../images/激活函数/identity.png')
plt.show()
```
